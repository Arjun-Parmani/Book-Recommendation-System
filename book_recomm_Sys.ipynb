## Popularity based recomm system

```python
rating_with_books=rating.merge(books, on='ISBN')
```

```python
rating_with_books.head()
```

```python
num_rating = rating_with_books.groupby('Book-Title').count()['Book-Rating'].reset_index()
```

```python
num_rating.rename(columns={'Book-Rating':'num_rating'},inplace=True)
```

```python
num_rating
```

```python
avg_rating = rating_with_books.groupby('Book-Title')['Book-Rating'].mean().reset_index()
```

```python
avg_rating.rename(columns={'Book-Rating':'avg_rating'},inplace=True)
avg_rating
```

```python
Popular_df = avg_rating.merge(num_rating,on='Book-Title')
```

```python
Popular_df
```

```python
Popular_df=Popular_df[Popular_df['num_rating']>=250].sort_values('avg_rating',ascending=False).head(50)
```

```python
Popular_df=Popular_df.merge(books,on='Book-Title').drop_duplicates('Book-Title')[['Book-Title','Book-Author','Year-Of-Publication','num_rating','avg_rating','Image-URL-M']]
```

```python
Popular_df.shape
```

## Collaborative approach

```python
x=rating_with_books.groupby('User-ID').count()['Book-Rating']>200
Knowledge_users = x[x].index

```

```python
Filtered_rating = rating_with_books[rating_with_books['User-ID'].isin(Knowledge_users)]
```

```python
Filtered_rating
```

```python
y=Filtered_rating.groupby('Book-Title').count()['Book-Rating']>=50
Books_Rating_morethanfifty = y[y].index
```

```python
Books_Rating_morethanfifty
```

```python
Final_Filtered_rating = Filtered_rating[Filtered_rating['Book-Title'].isin(Books_Rating_morethanfifty)]
```

```python
Final_Filtered_rating
```

```python
pt=Final_Filtered_rating.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
```

```python
pt.fillna(0,inplace=True)
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_scores = cosine_similarity(pt)
```

```python
similarity_scores.shape
```

```python
def recommend(book_name):
    # index fetch
    index = np.where(pt.index==book_name)[0][0]
    similar_items = sorted(list(enumerate(similarity_scores[index])),key=lambda x:x[1],reverse=True)[1:5]
    
    data = []
    for i in similar_items:
        item = []
        temp_df = books[books['Book-Title'] == pt.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-M'].values))
        
        data.append(item)
    
    return data
```

```python
recommend('1984')
```
